```python
import nltk
from nltk import FreqDist
```

```python
file1 = "teen.txt"
file2 = "adult40s.txt"
f1= open(file1,"r")
f2 = open(file2, "r")
teen= f1.read()
adult40s = f2.read()
```

```python
# Regular Expression to match non-alphabetic characters
import re
# this regular expression pattern matches any word that contains all non-alphabetical
#   lower-case characters [^a-z]+
# the beginning ^ and ending $ require the match to begin and end on a word boundary 
pattern = re.compile('^[^a-z]+$')
```

```python
# match the pattern on a text string with all special (non-alphabetical) characters
nonAlphaMatch = pattern.match('**')
#  if it matched, print a message
if nonAlphaMatch: print ('matched non-alphabetical')
```

```python
# function that takes a word and returns true if it consists only
#   of non-alphabetic characters  (assumes import re)
def alpha_filter(w):
  # pattern to match word of non-alphabetical characters
  pattern = re.compile('^[^a-z]+$')
  if (pattern.match(w)):
    return True
  else:
    return False
```

```python
# get a list of stopwords from nltk
stopwords = open('Smart.English.stop')
stoptokens = nltk.word_tokenize(stopwords.read())
stoptokens.append("'s")
stoptokens.append("n't")
stoptokens.append("'m")
stoptokens.append("'ll")
stoptokens.append("'re")
stoptokens.append("'ve")
stoptokens.append("'d")
```

```python
teentokens = nltk.word_tokenize(teen) 
teenwords = [w.lower( ) for w in teentokens] 
# apply the function to remove non-alphabetical words
alphateenwords = [w for w in teenwords if not alpha_filter(w)]
# further filter the words by leaving out any word in the stopwords
stoppedteenwords = [w for w in alphateenwords if not w in stoptokens]
```

```python
adult40stokens = nltk.word_tokenize(adult40s) 
adult40swords = [w.lower( ) for w in adult40stokens] 
# apply the function to remove non-alphabetical words 
alphaadultwords = [w for w in adult40swords if not alpha_filter(w)]
# further filter the words by leaving out any word in the stopwords
stoppedadultwords = [w for w in alphaadultwords if not w in stoptokens]
```

```python
teendist = FreqDist(stoppedteenwords)
teenitems = teendist.most_common(50)
for item in teenitems:
  print(item)
```

```python
adultdist = FreqDist(stoppedadultwords)
adultitems = adultdist.most_common(50)
for item in adultitems:
  print(item)
```

```python
# setup for bigrams and bigram measures
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
# create the bigram finder and score the bigrams by frequency
teenfinder = BigramCollocationFinder.from_words(teenwords)
teenfinder.apply_word_filter(alpha_filter)
teenfinder.apply_word_filter(lambda w: w in stoptokens)
teenfinder.apply_freq_filter(5)
teenfinder.apply_ngram_filter(lambda w1, w2: len(w1) < 2)

teenscored = teenfinder.score_ngrams(bigram_measures.raw_freq)
for bscore in teenscored[:50]:
    print (bscore)
```

```python
# create the bigram finder and score the bigrams by frequency
adultfinder = BigramCollocationFinder.from_words(adult40swords)
adultfinder.apply_word_filter(alpha_filter)
adultfinder.apply_word_filter(lambda w: w in stoptokens)
adultfinder.apply_freq_filter(5)
adultfinder.apply_ngram_filter(lambda w1, w2: len(w1) < 2)
adultscored = adultfinder.score_ngrams(bigram_measures.raw_freq)
for bscore in adultscored[:50]:
    print (bscore)
```

```python
teenfinder.apply_freq_filter(5)
teenscoredpmi = teenfinder.score_ngrams(bigram_measures.pmi)
for bscore in teenscoredpmi[:50]:
    print (bscore)
```

```python
adultfinder.apply_freq_filter(5)
adultscored = adultfinder.score_ngrams(bigram_measures.pmi)
for bscore in adultscored[:50]:
    print (bscore)
```
